```python
import random
import numpy as np
import matplotlib.pyplot as plt

import torch
import torchvision

from torchvision import datasets
from torchvision import transforms
```

```python
transform = transforms.Compose(
    [
        transforms.ToTensor(),
    ]
)
training_data = datasets.FashionMNIST(root="../data/FashionMINST", download=True, transform=transform)

for i in range(25):
    sample_idx = random.randint(0,len(training_data)-1)
    img,label = training_data[sample_idx]
    plt.subplot(5, 5, i + 1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(img.permute(1,2,0))
    plt.title(label)
plt.show()
```
