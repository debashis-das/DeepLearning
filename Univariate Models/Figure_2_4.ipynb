# Mixture of 2 Gaussians

```python
import jax.numpy as jnp
import jax
import seaborn as sns
import matplotlib.pyplot as plt
import os

import scipy
from scipy.stats import norm

data = dict()
mu = [0, 2]
sigma = [0.5, 0.5]
weights = [0.5, 0.5]
data = {
    "distributions": [norm(loc=mu[i], scale=sigma[i]) for i in range(2)],
    "weights": weights,
    "x": jnp.linspace(-2, 2 * mu[1], 600),
}

# plot first distribution
plt.figure()
ax = plt.gca()

y1 = data['distributions'][0].pdf(data['x'])*data['weights'][0]
y2 = data['distributions'][1].pdf(data['x'])*data['weights'][1]
# Plot the final distribution and its mean.
y = sum(data['weights'][i] * data['distributions'][i].pdf(data['x']) for i in range(2))

ax.plot(data['x'], y, color='black', linestyle='dashed', linewidth=2, label='$0.5\mathcal{N}(x|0,0.5) + 0.5\mathcal{N}(x|2,0.5)$')
ax.plot(data['x'], y1, color='red', linestyle='dashdot', linewidth=1, label='$0.5\mathcal{N}(x|0,0.5)$')
ax.plot(data['x'], y2, color='green', linestyle='dashdot', linewidth=1, label='$0.5\mathcal{N}(x|0,0.5)$')
plt.legend(bbox_to_anchor=(1, 1))
sns.despine()
```
