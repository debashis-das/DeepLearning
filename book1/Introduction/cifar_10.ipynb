```python
try:
    import torch
except ModuleNotFoundError:
    %pip install -qq torch
    import torch
try:
    import torchvision
except ModuleNotFoundError:
    %pip install -qq torchvision
    import torchvision
    
import torchvision.transforms as transforms
import torchvision.datasets as datasets
import PIL.Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
torch.random.seed()

transform = transforms.Compose(
    [
        transforms.ToTensor(),
    ]
)

cifar10 = datasets.CIFAR10('../data/CIFAR10', download=True, train=True, transform=transform)

class_names = ["plane", "car", "bird", "cat", "deer", "dog", "frog", "horse", "ship", "truck"]
for i in range(25):
    sample_idx = i
    img,label = cifar10[sample_idx]
    plt.subplot(5, 5, i + 1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(img.permute(1,2,0))
    plt.title(class_names[label])
plt.show()
```
